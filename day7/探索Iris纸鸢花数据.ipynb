```python
import pandas as pd
import numpy as np
```

```python
iris = pd.read_csv('../data_sets/day7/iris.csv',names = ['sepal_length','sepal_width', 'petal_length', 'petal_width', 'class'])
iris.head()
```

1. 数据框中有缺失值吗？

```python
iris.isnull().sum()
```

2. 将列petal_length的第10到19行设置为缺失值

```python
iris.iloc[10:20,2:3] = np.nan
iris.head(10)
```

3. 将缺失值全部替换为1.0

```python
iris.petal_length.fillna(1, inplace = True)
iris.head(10)
```

4. 删除列class

```python
del iris['class']
iris.head()
```

5. 将数据框前三行设置为缺失值

```python
iris.iloc[0:3 ,:] = np.nan
iris.head()
```

6. 删除有缺失值的行

```python
iris = iris.dropna(how='any')
iris.head()
```

7. 重新设置索引

```python
iris = iris.reset_index(drop = True)
iris.head()
```
